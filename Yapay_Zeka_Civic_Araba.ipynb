```python
import pandas as pd
import matplotlib as plt

df = pd.read_csv("/content/civic-yeni-data.csv")
print(df.shape)
```

#Veri Ön İşleme Adımları

# 1- Null Değerlerin Silinmesi

```python
# sadece title olarak girilen diğer değerleri olmayan veriler çıkartılıyor.
bos_index=df[df["year"].isnull()].index.tolist()

df=df.drop(index=bos_index)
df=df.reset_index(drop=True)
df.isnull().sum()
```

```python
# Engine power ve engine Capatiy verileri içerisinde null olarak atanan veriler indexleri ile bulunup silinme işlemi yapıldı.

bos_motor_gucu=df[df["enginePower"].isnull()].index.tolist()
df = df.drop(index=bos_motor_gucu)

bos_motor_kapasite=df[df["engineCapacity"].isnull()].index.tolist()
df= df.drop(index=bos_motor_kapasite)
df.isnull().sum()
```

```python
df["painted"]=df["painted"].fillna("yok")
df["changed"] = df["changed"].fillna("yok")
# Painted ve Changed columnları null olan rowları "yok" verisiyle doldurduk
df.head()
```

```python
df.isnull().sum()
# Artık Eksik verilerimiz yok
```

# 2- Features Seçimi

```python
# Genetik Algoritma kullanılarak Features seçimi işlemi yapılacak.
#Koveryans alalım daha sonrasında genetik algoritma ile features selection işlemi yapılarak bir kıyaslama yapalım.
# İlk olarak kolerasyon analizi ile features seçimi yapılım.

# mileage , year , transsmision, enginePower , color , model , engineCapacity, fuelType, paintedCount,changedCount
```

```python
df = df.drop(columns=["title","location"])
```

```python
df.head()
```

#3- Row verilerinin makinenin anlayacağı formata dönüştürme işlemleri

```python
# price String olarak yazılmış bunu float değere dönüştürmemiz lazım
def price_temizleme(price_str):
  if isinstance(price_str,str):
    temizlenmis_price = price_str.replace("TL","").replace(" ","").replace(".","")
    return float(temizlenmis_price)
  return None
df["price"] = df["price"].apply(price_temizleme)
```

```python
# EnginePower ve engineCapacity sutunlarını hazır hale getirdik.
def temizleme(val):
  if isinstance(val,str):
    temizlenmis_veri = val.lower().replace("cm3","").replace("cc","").replace("hp","").replace(" ","")
    return temizlenmis_veri
df["enginePower"] = df["enginePower"].apply(temizleme)
df["engineCapacity"]=df["engineCapacity"].apply(temizleme)
```

```python
df.head()
```

```python
# one-hot encoding işlemini manuel olarak ayırdık
all_parts = pd.concat([df["painted"], df["changed"]], ignore_index=True)

# "Yok" olmayanlardan parçaları ayır
unique_parts = set()

for val in all_parts.dropna():
    if val.lower().strip() == "yok":
        continue
    parts = [p.strip() for p in val.split(";")]
    unique_parts.update(parts)

# Liste haline getir
unique_parts = sorted(unique_parts)
```

```python
# parçaların üstünde değişiklik veya boyanmış ise 0 ve 1 değerleri giriliyor.
for part in unique_parts:
    col_name_changed = f"changed_{part}"
    col_name_painted = f"painted_{part}"

    df[col_name_changed] = df["changed"].apply(
        lambda x: 1 if isinstance(x, str) and part in x else 0
    )

    df[col_name_painted] = df["painted"].apply(
        lambda x: 1 if isinstance(x, str) and part in x else 0
    )
```

```python
df.head()
```

```python
df=df.drop(columns=["painted","changed"])
```

```python
df.head()
```

```python
df.to_csv("güncel_veri_seti_2",index=False)
```

```python
from sklearn.preprocessing import OneHotEncoder

# One-hot encoding yapılacak sütunlar
one_hot_cols = ["transmission", "model", "color", "fuelType"]

# Encoder nesnesi
encoder = OneHotEncoder(drop='first', sparse_output=False, dtype=int)

# Encode et ve sütun isimlerini al
encoded_array = encoder.fit_transform(df[one_hot_cols])
encoded_df = pd.DataFrame(encoded_array, columns=encoder.get_feature_names_out(one_hot_cols))

# Index'leri eşitle
encoded_df.index = df.index

# Orijinal sütunları drop et, yenileri birleştir
df = df.drop(columns=one_hot_cols)
df = pd.concat([df, encoded_df], axis=1)

df.head()
```

```python
df.to_csv("güncel_veri_seti_4",index=False)
```

```python
# bazı veriler 76-100 veya 1201-1300 gibi yazılmış bunların ortalama değeri alınarak güncelledik.
import numpy as np
def convert_range_to_mean(val):
    try:
        if isinstance(val, str) and "-" in val:
            parts = val.split("-")
            nums = [int(p.strip()) for p in parts]
            return sum(nums) / len(nums)
        else:
            return float(val)
    except:
        return np.nan

# Engine Power sütununu işle
df["enginePower"] = df["enginePower"].apply(convert_range_to_mean)

# Engine Capacity sütununu işle
df["engineCapacity"] = df["engineCapacity"].apply(convert_range_to_mean)
```

```python
# yukarıdaki işlemi yaptıktan sonra anlam veremediğim bir olay gerçekleşti.
# enginePower değerlerinde 10 adet null değer çıktı
# çok fazla olmadığı için veri setinden çıkardım.
df = df.dropna(subset=['enginePower'])
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error, r2_score

# year değerini yaş olarak yazalım.
df["age"] = 2025 - df["year"]
columns_to_drop = ["year"]
X = df.drop(columns=["price"] + columns_to_drop)
X["age"] = df["age"]
y = df["price"]
```

```python
price_threshold = y.quantile(0.99)
X = X[y < price_threshold]
y = y[y < price_threshold]
y_log = np.log1p(y)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y_log, test_size=0.2, random_state=42)
```

```python
# 8. Lineer regresyon modelini eğit
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
# 9. Tahmin yap
y_pred_log = model.predict(X_test)
y_pred = np.expm1(y_pred_log)
y_test_actual = np.expm1(y_test)
```

```python
# 10. Performansı değerlendir
mae = mean_absolute_error(y_test_actual, y_pred)
r2 = r2_score(y_test_actual, y_pred)

print(f"MAE: {mae:.2f} TL")
print(f"R² Score: {r2:.4f}")
```
